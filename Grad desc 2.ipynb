```python
import numpy as np
```

```python
def cost(points, m):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i,:5]
        y = points[i,5]
        s=0
        for j in range(len(m)):
            s=s+m[j]*x[j]
        total_cost += (1/M)*((y - s)**2)
    return total_cost
```

```python
def step_gradient(points, learning_rate, m):
    m_slope = np.zeros(len(m),dtype=float)
    n=len(m)
    new_m = m
    M = len(points)
    for i in range(M):
        x = points[i,:5]
        y = points[i, 5]
        for j in range(len(m_slope)):
            s=0
            for k in range(n):
                s=s+m[k]*x[k]
                
            m_slope[j]+=(-2/M)* (y -s)*x[j]
#         new_m =new_m - learning_rate*m_slope
       
#      print(m_slope)
    new_m = m - learning_rate*m_slope
#     print(new_m)
  
    return new_m
```

```python
def gd(points, learning_rate, num_iterations):
    m =np.zeros(5,dtype=int)
  
    for i in range(num_iterations):
        m= step_gradient(points, learning_rate, m)
        print(i, " Cost: ", cost(points, m))
    return m
```

```python
def run(data2):

    learning_rate =0.3
    num_iterations = 1000
    m= gd(data2, learning_rate, num_iterations)
    
    return m
```

```python
data = np.loadtxt("0000000000002419_training_ccpp_x_y_train.csv", delimiter=",")

data2=np.zeros((len(data),6),dtype=float)
data2[:,:4]=data[:,:4] #putting features
data2[:,4]=1            #putting column with all 1
data2[:,5]=data[:,4]   #putting output

from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
scaler.fit(data2[:,:4])
    
data2[:,:4]= scaler.transform(data2[:,:4])
m=run(data2)
```
