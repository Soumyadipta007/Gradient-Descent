```python
def cost(points, m):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i,:14]
        y = points[i,14]
        s=0
        for j in range(len(m)):
            s=s+m[j]*x[j]
        total_cost += (1/M)*((y - s)**2)
    return total_cost
```

```python
def step_gradient(points, learning_rate, m):
    m_slope = np.zeros(len(m),dtype=float)
    n=len(m)
    
    M = len(points)
    for i in range(M):
        x = points[i,:14]
        y = points[i, 14]
        for j in range(len(m_slope)):
            s=0
            for k in range(n):
                s=s+m[k]*x[k]
                
            m_slope[j]+=(-2/M)* (y -s)*x[j]
        
       
#     print(m_slope)
    new_m = m - learning_rate*m_slope
#     print(new_m)
  
    return new_m
```

```python
def gd(points, learning_rate, num_iterations):
    m =np.zeros(14,dtype=int)
  
    for i in range(num_iterations):
        m= step_gradient(points, learning_rate, m)
        print(i, " Cost: ", cost(points, m))
    return m
```
