```python
import numpy as np
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler()
```

```python
def Step_Gradient_Descent(x, y, learning_rate, m, c):
    size = x.shape[1]
    c_slope = 0
    m_slope = np.zeros(size)
    n = len(x)
    for i in range(n):
        x_ = x[i]
        y_ = y[i]
        eff = x_*m
        mx = np.sum(eff)
        slope = (-2/n)*(y_ - mx - c)
        for j in range(size):
            m_slope[j] += slope*x_[j]
        c_slope += slope
        new_m = m - learning_rate*m_slope
        new_c = c - learning_rate*c_slope
    return new_m, new_c
```

```python
def Cost(x, y, m, c):
    total_cost = 0
    n = len(x)
    for i in range(n):
        x_ = x[i]
        y_ = y[i]
        eff = x_*m
        mx = np.sum(eff)
        total_cost += (1/n)*((y_ - mx - c)**2)
    return total_cost
```

```python
def Gradient_Descent(x, y, learning_rate, number_of_iterations):
    size = x.shape[1]
    m = np.zeros(size)
    c = 0
    for i in range(number_of_iterations):
        m, c = Step_Gradient_Descent(x, y, learning_rate, m, c)
        print(i, Cost(x, y, m, c))
    return m, c
```
